# OpenCV (`cv2`) install note (Windows)

> Your current Jupyter kernel is using **Python 3.14**.

- `numpy<2` and `opencv-python` typically **do not provide wheels for Python 3.14** yet, so `pip` tries to compile from source and fails (needs Visual Studio C++ build tools).
- The simplest fix is to use **Python 3.10 or 3.11** for this project, then install `numpy<2` and `opencv-python<4.10` into that environment.

If you see build/compiler errors or `ImportError: numpy.core.multiarray failed to import`, switch the notebook kernel/interpreter to Python 3.10/3.11 and re-run the install cell below.

```python
import sys
import subprocess

print("Python:", sys.version)
print("Executable:", sys.executable)

# OpenCV + this TensorFlow Object Detection workflow is best supported on Python 3.10/3.11.
# On Python 3.13+ you often won't get prebuilt wheels for numpy/opencv yet, so pip tries (and fails) to compile from source.
if sys.version_info >= (3, 13):
    print("\n❗ Unsupported Python for this notebook:", sys.version.split()[0])
    print("\nFix:")
    print("1) Install Python 3.11 (64-bit) or 3.10")
    print("2) VS Code → Command Palette → Python: Select Interpreter → choose Python 3.11")
    print("3) VS Code → Command Palette → Jupyter: Select Kernel → choose the same Python 3.11")
    print("4) Restart the kernel, then rerun this cell")
else:
    # Install into *this* kernel's environment (avoids 'pip installs into a different Python').
    subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "pip"])
    subprocess.check_call([sys.executable, "-m", "pip", "install", "numpy<2.0", "opencv-python<4.10", "matplotlib"])

    import cv2
    print("cv2 version:", cv2.__version__)
```

```python
IMAGES_PATH = 'Tensorflow/workspace/images/collectedimages'
```

```python
labels = ['hello' , 'thanks' , 'yes' , 'no' , 'iloveyou']
number_imgs = 15
```

```python
import os
import time
import uuid

import cv2
import matplotlib.pyplot as plt
from IPython.display import clear_output

cap = cv2.VideoCapture(0)

for label in labels:
    label_path = os.path.join(IMAGES_PATH, label)
    os.makedirs(label_path, exist_ok=True)

    print(f"collecting images for {label}")
    time.sleep(5)

    for imgname in range(number_imgs):
        ret, frame = cap.read()
        if not ret:
            print("❌ Failed to read from webcam")
            break

        img_path = os.path.join(
            IMAGES_PATH,
            label,
            f"{label}.{uuid.uuid1()}.jpg"
        )

        cv2.imwrite(img_path, frame)

        # Show frame in Jupyter
        clear_output(wait=True)
        plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        plt.title(f"{label} - image {imgname+1}/{number_imgs}")
        plt.axis("off")
        plt.show()

        time.sleep(2)

cap.release()
```
